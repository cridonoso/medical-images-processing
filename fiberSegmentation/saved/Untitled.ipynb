```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('times.csv')
df.head()
names = [n for n in df['name']]
time = [n for n in df['time']]
df2 = pd.DataFrame(time, index=names)
```

```python
times = df['time']
total = times.sum()/60.
print total,'min'
df2.columns = ['time']
df2 = df2.sort_values('time', ascending=True)
```

```python
fig = plt.figure()
df_cum = df2.cumsum()
#a = fig.add_subplot(3, 2, 5)
ax = df_cum['time'].plot(kind='bar', figsize=(15,8), title='Time cumulative for each part',grid=True)
```
