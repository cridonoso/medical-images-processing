```python
import pickle
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from read_write_bundles import *
```

```python
rootdir = '.'
extensions = ('.pkl')
for subdir, dirs, files in os.walk(rootdir):
    for file in files:
        ext = os.path.splitext(file)[-1].lower()
        if ext in extensions:
            print file
            
file = open("atlas_UN_RIGHT_MNI.bundles.pkl",'rb')
object_file = pickle.load(file)

path2 = "../whole_brain_MNI_100k_21p.bundles"
whole_brain = read_bundle(path2)
```

```python
xline,yline,zline = [], [], []
for i in whole_brain[0:10000]:
    x,y,z = [],[],[]
    for j in i:
        x.append(j[0])
        y.append(j[1])
        z.append(j[2])
    xline.append(x)
    yline.append(y)
    zline.append(z)

xline2,yline2,zline2 = [], [], []
for i in object_file:
    x,y,z = [],[],[]
    for j in i:
        x.append(j[0])
        y.append(j[1])
        z.append(j[2])
    xline2.append(x)
    yline2.append(y)
    zline2.append(z)
```

```python
fig = plt.figure(figsize=(30,20))
ax = plt.axes(projection='3d')
for i,(x) in enumerate(xline):  
    ax.plot3D(xline[i], yline[i], zline[i],'gray')
for i,(x) in enumerate(xline2):  
    ax.plot3D(xline2[i], yline2[i], zline2[i],'red')
plt.title('AR_ANT_LEFT_MNI')
plt.show()
```
